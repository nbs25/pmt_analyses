```python
import numpy as np
```

```python
def GetDirectionalUnitVector(xi,xf):
    
    assert len(xi) == 3
    assert len(xi) == len(xf)
  
    xi = np.array(xi)
    xf = np.array(xf)
    
    unit = (xf - xi) / np.sqrt(np.dot(xf-xi,xf-xi))
        
    assert len(unit) == 3
    assert abs(np.dot(unit,unit)-1)< 0.01
    
    return unit

def GetMomentumMagnitude(E,m):
    
    assert E > m
    
    p = np.sqrt(E*E-m*m)
    
    return p

def GetMomentum(xi,xf,E,m):
    
    assert len(xi) == 3
    assert len(xi) == len(xf)
    
    p = GetMomentumMagnitude(E,m) * GetDirectionalUnitVector(xi,xf)
    
    assert len(p) == 3
    
    return p

def PrintHepevt_ParticleGun(_pdg, _xi, _xf, _E, _m, _t0):
    
    eventNumber = 0
    Nparticles  = 1
    
    print(eventNumber, Nparticles)
    
    status_code = 1
    pdg = _pdg
    
    mother1   = 0
    mother2   = 0
    daughter1 = 0
    daughter2 = 0
    
    p = GetMomentum(_xi,_xf,_E,_m)
    
    px = p[0]
    py = p[1]
    pz = p[2]
    
    x = _xi[0]
    y = _xi[1]
    z = _xi[2]
    
    t0 = _t0 # i think that this is actually important in this situation, but setting it to zero for the time being
    
    print(status_code, pdg, mother1, mother2, daughter1, daughter2, px, py, pz, _E, _m, x, y, z, t0)
```
